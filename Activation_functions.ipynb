```python
import numpy as np
import matplotlib.pyplot as plt 


def sigmoid(X): 
  return 1 / (1 + np.exp(-X))
  
X = np.linspace(-10, 10)
plt.plot(X, sigmoid(X))
plt.axis('tight')
plt.title('Activation function : Sigmoid')
plt.show()
```

```python
def relu(x):
  x1 = []
  for i in x:
    if i < 0:
      x1.append(0)
    else:
      x1.append(i)

  return x1
  
x = np.linspace(-10, 10)
plt.plot(x, relu(x))
plt.axis('tight')
plt.show()
```
